```python
import  pandas as pd

df=pd.read_csv("car (1).data",names=['buying','maint','doors','person','lug_boot','safty','class'])
df.head()
```

```python
k={'2':2,'3':3,'4':4,'5':5,'more':6,'5more':6}
df['doors']=df['doors'].map(k)
df['person']=df['person'].map(k)
```

```python
d={'vhigh':4,'high':3,'med':2,'low':1}
df['buying']=df['buying'].map(d)
df['maint']=df['maint'].map(d)
df['lug_boot']=df['lug_boot'].map({'small':1,'med':2,'big':3})
df['safty']=df['safty'].map({'low':1,'med':2,'high':3})
df['class']=df['class'].map({'unacc':4,'acc':3,'good':2,'vgood':1})
df.head(15)
```

```python
features=['buying','maint','doors','person','lug_boot','safty']
labels=['class']
```

```python
import sklearn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import numpy as np

feat=np.array(df[features].values)
labl=np.array(df[labels].values)

scaler=preprocessing.StandardScaler()
feat=scaler.fit_transform(feat)

x_train,x_test,y_train,y_test =train_test_split(feat,labl,test_size=0.2)
```

```python
from sklearn import neighbors

knn_=neighbors.KNeighborsClassifier(n_neighbors=15)

knn_.fit(x_train,y_train)
knn_.score(x_test,y_test)
```

```python
predt=knn_.predict(x_test)
for i in range(len(predt)):
    if(predt[i]!=y_test[i][0]):
        print(predt[i]," ",scaler.inverse_transform(x_test[i])," ",y_test[i][0])
```
